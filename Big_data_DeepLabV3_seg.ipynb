Load the images and their masks using their path from azure and applying some modifier as pre process

```python
# Setting some parameters
im_width = 256
im_height = 256

images = np.zeros(((len(id_images)), im_height, im_width), dtype=np.float32)
masks = np.zeros(((len(id_masks)), im_height, im_width), dtype=np.float32)


for i, file_name in tqdm_notebook(enumerate(id_images), total=len(id_images)):
    # Load images
    blob_client = container_client.get_blob_client(file_name)
    data = blob_client.download_blob().content_as_bytes()
    arr = np.frombuffer(data, np.uint8)
    img = cv2.imdecode(arr, cv2.IMREAD_GRAYSCALE)

    #pre process
    #Gaussian blur
    img = cv2.GaussianBlur(img,(3,3),cv2.BORDER_DEFAULT)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
    #morpho gradient
    image = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
    image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
    #egalization
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    img_clahe = clahe.apply(image)
    img = resize(img_clahe, (im_width, im_height), mode='constant', preserve_range=True)
    images[i] = img/255


for i, file_name in tqdm_notebook(enumerate(id_masks), total=len(id_masks)):
    # Load images
    blob_client = container_client.get_blob_client(file_name)
    data = blob_client.download_blob().content_as_bytes()
    arr = np.frombuffer(data, np.uint8)
    img = cv2.imdecode(arr, cv2.IMREAD_GRAYSCALE)
    img = resize(img, (im_width, im_height), mode='constant', preserve_range=True)
    masks[i] = img/255

```

#MODEL TRAINING
* DeepLabV3

```python
def get_deeplabv3(input_img, n_filters = 16, dropout = 0.1, batchnorm = True):
    """Function to define the deeplabv3 Model"""
    # feature extraction layers

    x = Conv2D(n_filters * 2, 3, activation="relu", padding="same")(input_img)
    x = BatchNormalization()(x)
    x = Conv2D(n_filters * 2, 3, activation="relu", padding="same")(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D()(x)
    x = Dropout(dropout)(x)


    x = Conv2D(n_filters * 4, 3, activation="relu", padding="same")(x)
    x = BatchNormalization()(x)
    x = Conv2D(n_filters * 4, 3, activation="relu", padding="same")(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D()(x)
    x = Dropout(dropout)(x)

    x = Conv2D(n_filters * 8, 3, activation="relu", padding="same")(x)
    x = BatchNormalization()(x)
    x = Conv2D(n_filters * 8, 3, activation="relu", padding="same")(x)
    x = BatchNormalization()(x)
    x = Conv2D(n_filters * 8, 3, activation="relu", padding="same")(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D()(x)
    x = Dropout(dropout)(x)
    
    # ASPP module
    x1 = Conv2D(n_filters * 8, 1, activation="relu", padding="same")(x)
    x1 = BatchNormalization()(x1)

    x2 = Conv2D(n_filters * 8, 3, dilation_rate=6, activation="relu", padding="same")(x)
    x2 = BatchNormalization()(x2)

    x3 = Conv2D(n_filters * 8, 3, dilation_rate=12, activation="relu", padding="same")(x)
    x3 = BatchNormalization()(x3)

    x4 = Conv2D(n_filters * 8, 3, dilation_rate=18, activation="relu", padding="same")(x)
    x4 = BatchNormalization()(x4)

    x = concatenate([x1, x2, x3, x4])
    x = Conv2D(n_filters * 8, 1, activation="relu", padding="same")(x)
    x = BatchNormalization()(x)

    # Decoder layers
    
    x = Conv2DTranspose(n_filters * 8, 3, strides=2, padding="same")(x)
    x = BatchNormalization()(x)
    x = Conv2DTranspose(n_filters * 4, 3, strides=2, padding="same")(x)
    x = BatchNormalization()(x)
    x = Conv2DTranspose(n_filters * 2, 3, strides=2, padding="same")(x)
    x = BatchNormalization()(x)

    outputs = Conv2D(1, (1, 1), activation='sigmoid')(x)
    model = Model(inputs=[input_img], outputs=[outputs])
    return model
```

Some various metrics to see the efficience

```python
def iou(y_true, y_pred):
    intersection = np.logical_and(y_true, y_pred)
    union = np.logical_or(y_true, y_pred)
    iou_score = np.sum(intersection) / np.sum(union)
    return iou_score

def dice_coef(y_true, y_pred):
    intersection = np.sum(y_true * y_pred)
    dice_score = (2. * intersection) / (np.sum(y_true) + np.sum(y_pred))
    return dice_score
```
